```sql

CREATE CATALOG IF NOT EXISTS instacart
COMMENT "catalog for instacart";
USE CATALOG instacart;
    
CREATE SCHEMA IF NOT EXISTS instacart_schema
COMMENT "schema for instacart";

create schema if not exists bronze_schema
COMMENT "schema for bronze tables";
create schema if not exists silver_schema
COMMENT "schema for silver tables";
create schema if not exists gold_schema
COMMENT "schema for gold tables"
```
