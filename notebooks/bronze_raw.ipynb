```sql
CREATE TABLE IF NOT EXISTS instacart.bronze_schema.product_t 
comment "Product table raw data" AS
SELECT * FROM read_files(
  '/Volumes/instacart/instacart_schema/instamart_volume/products.csv',
  format => 'csv', delimiter => ',',
  header => true,
  escape => '\\',
  quote => '"',
  inferSchema => true
)
;
```
